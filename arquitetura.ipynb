```python
from graphviz import Digraph
from IPython.display import Image, display

# Criação do objeto Digraph
dot = Digraph()

# Adiciona um título ao fluxograma indicando que é uma arquitetura local
dot.attr(label='Arquitetura Local', labelloc='t', fontsize='20', fontcolor='black', style='bold')
dot.attr(fontsize='12', fontcolor='black')

# Adiciona nós principais ao fluxograma
dot.node('A', 'Coleta de Dados\nAPI/InduSoft', shape='ellipse', style='filled', fillcolor='lightblue', fontcolor='black')
dot.node('B', 'Preprocessamento de Dados\nRemoção de Ruído', shape='box', style='filled', fillcolor='lightgreen', fontcolor='black')
dot.node('C', 'Normalização de Dados\nEscalonamento', shape='diamond', style='filled', fillcolor='lightyellow', fontcolor='black')
dot.node('D', 'Detecção de Outliers\nAlgoritmo de ML', shape='trapezium', style='filled', fillcolor='lightcoral', fontcolor='black')
dot.node('E', 'Armazenamento de Dados\nBanco de Dados Local', shape='parallelogram', style='filled', fillcolor='lightgray', fontcolor='black')
dot.node('F', 'Visualização\nPainel de Análise', shape='octagon', style='filled', fillcolor='lightpink', fontcolor='black')
dot.node('G', 'Ação Corretiva\nAutomatização de Processos', shape='hexagon', style='filled', fillcolor='lightgoldenrodyellow', fontcolor='black')

# Adiciona nós de falha para todas as etapas
dot.node('A_fail', 'Falha na Coleta de Dados\nMensagem: Erro na API/InduSoft', shape='box', style='filled', fillcolor='red', fontcolor='white')
dot.node('B_fail', 'Falha no Preprocessamento\nMensagem: Erro na Remoção de Ruído', shape='box', style='filled', fillcolor='red', fontcolor='white')
dot.node('C_fail', 'Falha na Normalização\nMensagem: Erro no Escalonamento', shape='box', style='filled', fillcolor='red', fontcolor='white')
dot.node('D_fail', 'Falha na Detecção de Outliers\nMensagem: Erro no Algoritmo de ML', shape='box', style='filled', fillcolor='red', fontcolor='white')
dot.node('E_fail', 'Falha no Armazenamento\nMensagem: Erro no Banco de Dados', shape='box', style='filled', fillcolor='red', fontcolor='white')
dot.node('F_fail', 'Falha na Visualização\nMensagem: Erro na Criação de Relatórios/Dashboards', shape='box', style='filled', fillcolor='red', fontcolor='white')
dot.node('G_fail', 'Falha na Ação Corretiva\nMensagem: Erro na Automatização', shape='box', style='filled', fillcolor='red', fontcolor='white')

# Subdivide os nós principais em subetapas
# Coleta de Dados
dot.node('A1', 'API\nCaptura de Dados', shape='box', style='filled', fillcolor='lightblue', fontcolor='black')
dot.node('A2', 'InduSoft\nIntegração', shape='box', style='filled', fillcolor='lightblue', fontcolor='black')

# Preprocessamento de Dados
dot.node('B1', 'Tratamento de Dados\nRemoção de Nulos', shape='box', style='filled', fillcolor='lightgreen', fontcolor='black')
dot.node('B2', 'Limpeza de Dados\nFormatação', shape='box', style='filled', fillcolor='lightgreen', fontcolor='black')

# Normalização de Dados
dot.node('C1', 'Escalonamento\nMin-Max', shape='box', style='filled', fillcolor='lightyellow', fontcolor='black')
dot.node('C2', 'Normalização\nZ-score', shape='box', style='filled', fillcolor='lightyellow', fontcolor='black')

# Detecção de Outliers
dot.node('D1', 'Método de ML\nIsolamento de Florestas', shape='box', style='filled', fillcolor='lightcoral', fontcolor='black')
dot.node('D2', 'Análise Estatística\nDesvio Padrão', shape='box', style='filled', fillcolor='lightcoral', fontcolor='black')

# Armazenamento de Dados
dot.node('E1', 'MySQL\nPersistência', shape='box', style='filled', fillcolor='lightgray', fontcolor='black')
dot.node('E2', 'SQLite\nBackup', shape='box', style='filled', fillcolor='lightgray', fontcolor='black')

# Visualização
dot.node('F1', 'Criação de Relatórios\nCustomização', shape='box', style='filled', fillcolor='lightpink', fontcolor='black')
dot.node('F2', 'Dashboards\nInteratividade', shape='box', style='filled', fillcolor='lightpink', fontcolor='black')

# Ação Corretiva
dot.node('G1', 'Automatização\nProcessos Repetitivos', shape='box', style='filled', fillcolor='lightgoldenrodyellow', fontcolor='black')
dot.node('G2', 'Correção de Dados\nAjustes Automáticos', shape='box', style='filled', fillcolor='lightgoldenrodyellow', fontcolor='black')

# Adiciona conexões entre os nós principais e suas subetapas
dot.edge('A', 'A1')
dot.edge('A', 'A2')
dot.edge('B', 'B1')
dot.edge('B', 'B2')
dot.edge('C', 'C1')
dot.edge('C', 'C2')
dot.edge('D', 'D1')
dot.edge('D', 'D2')
dot.edge('E', 'E1')
dot.edge('E', 'E2')
dot.edge('F', 'F1')
dot.edge('F', 'F2')
dot.edge('G', 'G1')
dot.edge('G', 'G2')

# Adiciona conexões entre as subetapas e as etapas principais
dot.edge('A1', 'B')
dot.edge('A2', 'B')
dot.edge('B1', 'C')
dot.edge('B2', 'C')
dot.edge('C1', 'D')
dot.edge('C2', 'D')
dot.edge('D1', 'E')
dot.edge('D2', 'E')
dot.edge('E1', 'F')
dot.edge('E2', 'F')

# Adiciona conexões entre as etapas principais e seus nós de falha
dot.edge('A', 'A_fail', color='red', style='dashed')
dot.edge('B', 'B_fail', color='red', style='dashed')
dot.edge('C', 'C_fail', color='red', style='dashed')
dot.edge('D', 'D_fail', color='red', style='dashed')
dot.edge('E', 'E_fail', color='red', style='dashed')
dot.edge('F', 'F_fail', color='red', style='dashed')
dot.edge('G', 'G_fail', color='red', style='dashed')

# Adiciona conexões entre os nós de falha e as ações corretivas
dot.edge('A_fail', 'G1')
dot.edge('B_fail', 'G1')
dot.edge('C_fail', 'G1')
dot.edge('D_fail', 'G1')
dot.edge('E_fail', 'G1')
dot.edge('F_fail', 'G1')
dot.edge('G_fail', 'G1')

# Renderiza e exibe o fluxograma
dot.format = 'png'
file_path = dot.render(filename='fluxograma_arquitetura_local_com_falhas_completas', directory='.', cleanup=True)

# Exibe a imagem do fluxograma
display(Image(filename=file_path))
```
